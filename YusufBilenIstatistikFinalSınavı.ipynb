```python
import pandas as pd

# Veri dosyasını oku
df = pd.read_csv('ramen-ratings.csv')

# Veri setinin başını göster
df.head()
```

```python
# Sonucu ekrana yazdır
top_10_restaurants
```

# Soru 2;

```python
import pandas as pd

# Verileri yükleyin
df = pd.read_csv("winemag-data-130k-v2.csv")
df
```

```python
# Türkiye'den şarapları seç
sarap = df[df['country'] == 'Turkey']
sarap
```

```python

# Puanlarına göre sırala (en yüksek puandan en düşük puana)
sorted_wines = sarap.sort_values(by='points', ascending=False)
sorted_wines
```

```python
# Fiyat aralığı uygun olan şarapları seç
affordable_wines = sorted_wines[sorted_wines['price'] < 50]
affordable_wines
```

```python
# İlk 7 şarabı seç
top_7_wines = affordable_wines.head(7)

# Sonucu ekrana yazdır
top_7_wines[['title', 'points', 'price']]
```

# Soru 3;

```python
import pandas as pd
#Öncelikle bu kodda verilerine ulaşmak istediğimiz dosyanın kodunu yazıyoruz
df = pd.read_csv("fifa.csv")
df
```

```python
#Bu kodda takımları tanıttık
teams = ["ARG", "BRA", "ESP", "FRA", "GER", "ITA"]
subset = df[teams]
subset
```

```python
#Bu kodda takımlarn sıralamasınnı yaptık
subset = subset[::-1]
latest_ranking = subset.loc[subset.index[0]]
latest_ranking
```

```python
#Bu kodda ortalamaları yüksek olan iki ülkeyi seçtik
top_teams = latest_ranking.nlargest(2)
top_teams
```

```python
#Bu kodda da bu takımları tanımladık
team1 = top_teams.index[0]
team2 = top_teams.index[1]

print("Seçilen takımlar:", team1, "ve", team2)
```

# Soru 4;

```python
import pandas as pd

# Verileri oku
df = pd.read_csv("flight_delays.csv")
df
```

```python

# Sadece gecikme verilerini içeren sütunları seç
selected_columns = ['Month', 'AA', 'AS', 'B6', 'DL', 'EV', 'F9', 'HA', 'MQ', 'NK', 'OO', 'UA', 'US', 'VX', 'WN']
df_selected = df[selected_columns]
df_selected
```

```python
# Havayolu şirketlerinin ortalama gecikme sürelerini hesapla
average_delays = df_selected.mean()
average_delays
```

```python
# Geçikmeleri küçükten büyüğe sırala
sorted_delays = average_delays.sort_values()
sorted_delays
```

```python
# En düşük geçikmeye sahip olan havayolu şirketini seç
selected_airline = sorted_delays.index[0]
selected_airline
```
